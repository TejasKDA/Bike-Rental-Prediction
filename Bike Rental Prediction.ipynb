# Dummy Variables

```python
#Season Dummy Creation
seasondummy=pd.get_dummies(BikeRental['season']).astype(int)
seasondic={1:'spring', 2:'summer', 3:'fall', 4:'winter'}
seasondummy.rename(seasondic, axis=1 ,inplace=True)
seasondummy.drop(['spring'],axis=1,inplace=True)
seasondummy.head()
```

```python
#Weather Dummy Creation
WeathersitDummy=pd.get_dummies(BikeRental['weathersit']).astype(int)
WeathersitDummy.head()
weatherdic={1:'Partly cloudy', 2:'Mist and Cloudy', 3:'Light Rain'}
WeathersitDummy.rename(weatherdic, axis=1 ,inplace=True)
WeathersitDummy.drop(['Partly cloudy'],axis=1,inplace=True)
WeathersitDummy.head()
```

```python
#Month dummy creation
monthDummy=pd.get_dummies(BikeRental['mnth']).astype(int)
monthDummy.head()
monthdic={1:'Jan', 2:'Feb', 3:'Mar',4:'Apr', 5:'May', 6:'Jun',7:'Jul', 8:'Aug', 9:'Sep',10:'Oct', 11:'Nov', 12:'Dec'}
monthDummy.rename(monthdic, axis=1 ,inplace=True)
monthDummy.drop(['Jan'],axis=1,inplace=True)
monthDummy.head()
```

```python
#Weekday Dummy Creation
WeeekdayDummy=pd.get_dummies(BikeRental['weekday']).astype(int)
WeeekdayDummy.head()
weekdaydic={0:'Mon', 1:'Tue', 2:'Wed',3:'Thurs',4:'Fri', 5:'Sat', 6:'Sun'}
WeeekdayDummy.rename(weekdaydic, axis=1 ,inplace=True)
WeeekdayDummy.drop(['Mon'],axis=1,inplace=True)
WeeekdayDummy.head()
```

```python
BikeRental=pd.concat([BikeRental,seasondummy, WeathersitDummy,WeeekdayDummy,monthDummy], axis = 1)
BikeRental.head()
```

```python
#Drop the columns for which dummy variables are created and are repetative
BikeRental.drop(['season','weathersit','weekday','mnth'],axis=1,inplace=True)
```

```python
BikeRental.shape
```

# Splitting the Data into Training and Testing Sets


```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
bikerental_train, bikerental_test = train_test_split(BikeRental, train_size = 0.7, test_size = 0.3, random_state = 100)
```

# Rescaling Features

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['temp','atemp','hum','windspeed','cnt']
#fit- learns min and max
#transform apply minmax normalisation-(x-xmin/xmax-xmin)
bikerental_train[num_vars] = scaler.fit_transform(bikerental_train[num_vars])
```

# Divide the data in to X and Y Sets

```python
y_train = bikerental_train.pop('cnt')
X_train = bikerental_train
```

# RFE - Recursive Feature elimination

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#Run RFE for for 15 Features
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm,n_features_to_select=15)             
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[~rfe.support_]
```

```python
#Get the data for which the RFESupport is True 
col = X_train.columns[rfe.support_]
col
```

# Staistics model for detailed analysis

```python
import statsmodels.api as sm  
X_train_rfe = X_train[col]
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
#execute the linear model
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
#Summary of Statistics
print(lm.summary())
```

```python
#Check the VIF inorder to check the multi collinearity among independent variables
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Drop the columns first based on High pvalue and High VIF Value
#Next Drop the columns first based on High pvalue and Low VIF Value
#Next Drop the columns first based on Low pvalue and High VIF Value


#dropping "Sat" as it hugh p valueof 0.867 and is less significant
X_train_new = X_train_rfe.drop(['Sat'], axis=1)
```
